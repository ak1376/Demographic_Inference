```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load in the results from the preprocessing
with open("/sietch_colab/akapoor/Demographic_Inference/experiments/bigger_analysis/preprocessing_results_obj.pkl", "rb") as f:
    results = pickle.load(f)
```

```python
results['param_names']
```

```python
for i in np.arange(len(results['param_names'])):
    plt.figure()
    plt.hist(results['training']['predictions'][:,0,i])
    plt.suptitle(f'Dadi Results for {results["param_names"][i]}')
    plt.title(f"Min Value: {np.min(results['training']['predictions'][:,0,i]):.2f}, Max Value: {np.max(results['training']['predictions'][:,0,i]):.2f}")
    plt.show()
```

```python
for i in np.arange(len(results['param_names'])):
    plt.figure()
    plt.hist(results['training']['predictions'][:,1,i])
    plt.suptitle(f'Moments Results for {results["param_names"][i]}')
    plt.title(f"Min Value: {np.min(results['training']['predictions'][:,1,i]):.2f}, Max Value: {np.max(results['training']['predictions'][:,1,i]):.2f}")
    plt.show()
```
