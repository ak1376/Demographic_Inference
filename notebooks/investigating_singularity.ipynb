```python
import matplotlib.pyplot as plt

# Create a 5x3 grid of subplots
fig, axes = plt.subplots(5, 3, figsize=(15, 10))  # Adjust `figsize` as needed

# Flatten the axes array for easy indexing
axes = axes.flatten()

for i in range(len(data['stats'][0])):  # 15 statistics
    mean_list_stat = []
    for j in range(len(data['means']) - 1):  # j indexes the recombination bin number
        mean_list_stat.append(data['means'][j][i])
    
    # Plot on the i-th subplot in the grid
    axes[i].plot(mean_list_stat)
    axes[i].set_title(f"Statistic: {data['stats'][0][i]}")

# Hide any unused subplots if there are fewer than 15
for k in range(i + 1, len(axes)):
    axes[k].axis('off')

plt.tight_layout()  # Adjust spacing
plt.show()
```
